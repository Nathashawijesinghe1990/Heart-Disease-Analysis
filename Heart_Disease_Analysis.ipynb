## Data Transformation

The chol feature has an outlier which is very high and could lead to make a biased result. Therefore we will replace it with the mean value of the feature.

```python
# outliers of serum cholestorol
sns.boxplot(x=df['chol'])
```

```python
median= df['chol'].median()
```

```python
median
```

```python
outliers = (df['chol']> 370)
df[outliers]= np.nan
df['chol'].fillna(median, inplace= True)
```
